### 198. Fina a value in a BST

```python
class BST:
    
    #Function to search a node in BST.
    def search(self, node, x):
        #code here
        if node == None:
            return 0
        elif node.data == x:
            return 1
        elif x > node.data:
            return self.search(node.right,x)
        else:
            return self.search(node.left,x)
        
# Time comp: O(N)
# Auxiliary space: O(1)
```

### 199. Delete a node from BST

```python
def deleteNode(root, X):
    
    # If root is None, just return None
    if root == None:
        return None
        
    # Recursive call to find element
    if x > root.data:
        root.right = deleteNode(root.right, X)
        return root           # Don't forgot to return root from here
    elif x < root.data:
        root.left = deleteNode(root.left, X)
        return root           # Don't forgot to return root from here
    
    # If key has been found then only execution will come at this point
    
    # If found element is leaf then just return none (which indicate to delete that node)
    if root.left == None and root.right == None:
        return None
    # If only right child is there then return that child after deleting current node
    elif root.left == None:
        temp = root.right
        root == None
        return temp
    # If only left child is there then return that child after deleting current node
    elif root.right == None:
        temp = root.left
        root == None
        return temp
    # Now this code is for the case where two children is available
    else:
        # Find inorder successor
        prev = root
        curr = root.right
        while curr.left != None:
            prev = curr
            curr = curr.left
            
        root.data = curr.data

        # If that successor itself is right child of node then update right node of root
        if prev == root:
            prev.right = curr.right
        else:
            # else update left of prev node
            prev.left = curr.right
            
        # and return current root
        return root
    
# Time comp: O(N)
# Auxiliary Space: O(1)    (Recursion stack: O(N))
```

### 201. Find inorder successor and inorder predecessor in a BST 

```python
def inorderpre(root,key):
    pre = None
    while root:
        if key > root.key:
            pre = root
            root = root.right
        else: 
            root = root.left
    return pre

def inordersuc(root,key):
    suc = None
    while root:
        if key >= root.key:
            root = root.right
        else:
            suc = root
            root = root.left
    return suc

def findPreSuc(root, pre, suc, key):
    pre[0] = inorderpre(root,key)
    suc[0] = inordersuc(root,key)
    
# Time comp: O(N)
# Auxiliary space: O(1)
```

### 204. Lowest Common Ancestor in a BST

```python
def findPath(root,n,path):
    if root == None:
        return
    
    path.append(root)
    
    if root.data == n:
        return
    else:
        if n > root.data:
            findPath(root.right,n,path)
        else:
            findPath(root.left,n,path)

def LCA(root, n1, n2):
    path1 = []
    path2 = []
    
    findPath(root,n1,path1)
    findPath(root,n2,path2)
    
    if len(path1) == 0 or len(path2) == 0:
        return False

    prev = path1[0]
    n = min(len(path1),len(path2))
    for i in range(n):
        if len(path2) <= i or path1[i].data != path2[i].data:
            break
        prev = path1[i]
        
    return prev

# Time comp:O(N)
# Space comp: O(N)   (Because of two stack we have created)
```

### 205. Construct BST from preorder traversal

```python
class Solution:
    def findPartisan(self,preorder):
        x = preorder[0]
        for i in range(1,len(preorder)):
            if preorder[i] > x:
                return i
        return len(preorder)
    
    def construct(self,preorder):
        if preorder == []:
            return None
        if len(preorder) == 1:
            root = TreeNode(preorder[0])
            return root
        
        
        root = TreeNode(preorder[0])
        x = self.findPartisan(preorder)
        root.left = self.construct(preorder[1:x])
        root.right = self.construct(preorder[x:])
        return root
    
    def bstFromPreorder(self, preorder):
        return self.construct(preorder)
    
# Time comp:O(N^2)
# Auxiliary Space: O(1)
```

### 206. Binary Tree to BST

```python
"""
So the idea here is, make inorder traversal list first and sort it as inorder of BST are always sorted.
After that again traverse tree and update value in inorder one by one from the inorder array.
"""

class Solution:
    def __init__(self):
        self.key = 0
    
    def inorder(self,root,ans):
        if root == None:
            return
        
        self.inorder(root.left,ans)
        ans.append(root.data)
        self.inorder(root.right,ans)
    
    def update(self,root,ans):
        if root == None:
            return root
            
        if self.key > len(ans):
            return
        
        self.update(root.left,ans)
        
        root.data = ans[self.key]
        self.key += 1
        
        self.update(root.right,ans)
        
        return root
    
    def binaryTreeToBST(self, root):
        inorder = []
        self.inorder(root,inorder)
        inorder.sort()
        return self.update(root,inorder)
    
# Time comp: O(N log N)
# Auxiliary Space: O(N)
```

### 211. Count pairs from 2 BST whose sum is equal to given value "X"

```python
"""
Traverse first trees and make hash map of it
Traverse second trees and check whether its count value present in hash map or not
If its present then increment counter
"""

class Solution:
    def __init__(self):
        self.hash_map = {}
        self.count = 0
        
    def makeMap(self,root):
        if root == None:
            return
        
        self.makeMap(root.left)
        self.hash_map[root.data] = 1
        self.makeMap(root.right)

    def countPair(self,root,x):
        if root == None:
            return
        
        temp = x - root.data
        if temp in self.hash_map:
            self.count += 1
        
        self.countPair(root.left,x)
        self.countPair(root.right,x)
        
    def countPairs(self, root1, root2, x): 
        self.makeMap(root1)
        self.countPair(root2,x)
        return self.count
    
# Time comp:O(N+M)
# Space comp:O(N)  Since we are storing values in hash map
```

### 213. Count BST ndoes that lie in a given range

### 216. Preorder to BST

```python
"""
First node of the pre order is always a root node of the tree.
Based on this property, make first node a root node and then find the next bigger element than root node
lets say its index is 'i'
Left side of i will be the left subtree of the tree, while right side of the tree (including i) will be the RST
"""

def post_order(pre, size):
    if len(pre) == 0:
        return None
    
    root = Node(pre[0])
    if len(pre) == 1:
        return root
        
    i = 1
    while i < len(pre) and pre[i] < pre[0]:
        i += 1
    
    root.left = post_order(pre[1:i],size)
    root.right = post_order(pre[i:],size)
    
    return root

# Time comp:O(N^2)
# Space comp:O(N)   (Due to recursion)
```

### 217. Check whether BST contains Dead end

```python
"""
Here we need to check for each leaf node(i), whether i-1 and i+1 present in tree or not. 
If they are present, means there is dead end else not

The exception is 1. If 1 is at the leaf then only check whether 2 is present in tree or not, as we can't insert 0

"""

hash_map = {}

def makeHash(root):
    global hash_map
    
    if root == None:
        return
    
    if root.left == None and root.right == None:
        hash_map[root.data] = 1
    else:
        hash_map[root.data] = 0    
    
    if root.left:
        makeHash(root.left)
    
    if root.right:
        makeHash(root.right)

def isdeadEnd(root):
    global hash_map
    makeHash(root)
    
    for i in hash_map:
        if i == 1 and hash_map[i] == 1 and 2 in hash_map:
            hash_map = {}
            return True
            
        if i != 1 and hash_map[i] == 1 and i-1 in hash_map and i+1 in hash_map:
            hash_map = {}
            return True
    hash_map = {}
    return False

# Time comp:O(N)
# Space comp:O(N)   (As we used hash map)  (Recursive stack:O(N))
```

```python
"""
First of all, it is given that it is a BST and nodes are greater than zero, 
root node can be in the range [1, ∞] and if root val is say, val, 
then left sub-tree can have the value in the range [1, val-1] and right sub-tree the value in range [val+1, ∞]. 
we need to traverse recursively.
when the the min and max value of range coincided it means that we cannot add any node further in the tree.
"""

def checkDeadEnd(root,mn,mx):
    if root == None:
        return False   # Means no dead end
        
    if mn == mx:
        return True   # Means a dead end
        
    return checkDeadEnd(root.left,mn,root.data-1) or checkDeadEnd(root.right,root.data+1,mx)
    
def isdeadEnd(root):
    return checkDeadEnd(root,1,float('inf'))

# Time comp:O(N)
# Space comp:O(1)    (Recursive stack:O(N))
```

### 219. Flatten BST to sorted list

```python
new_root = None
prev = None

def inorder(root):
    global new_root
    global prev
    if root == None:
        return
    
    inorder(root.left)
    
    if prev == None:
        new_root = root
        prev = root
    else:
        prev.left = None
        prev.right = root
        prev = root
    
    inorder(root.right)

def flatten(root):
    global new_root
    global prev
    new_root = None
    prev = None
    inorder(root)
    return new_root

# Time comp:O(N)
# Space comp:O(1)    (Recursive stack:O(N))
```

```python
# Brute force solution
# For each element, find next greater element at right side and consider it as mid
# So all the element before mid should be less than current element 
# and all the element to the right side should be greater than current element

class Solution:
    
    # Find next greater element
    def findNextMid(self,arr,key):
        for i in range(len(arr)):
            if arr[i] > key:
                return i
        return -1
    
    # Check whether left sub tree is valid or not
    def isValidRight(self,arr,key):
        for i in arr:
            if i < key:
                return False
        return True
        
    # Check whether right sub tree is valid or not
    def isValidLeft(self,arr,key):
        for i in arr:
            if i > key:
                return False
        return True
    
    def canRepresentBST(self, arr, N):
        for i in range(N):
            mid = self.findNextMid(arr[i+1:],arr[i])
            
            if mid == -1:
                mid = N
            else:
                # did this because we have call findNextMid with array[i+1:], so we need to add i in result
                mid += i+1
                
            if mid != i+1:
                x = self.isValidLeft(arr[i+1:mid],arr[i])
                if x == False:
                    return 0
            
            if mid != N:
                x = self.isValidRight(arr[mid:],arr[i])
                if x == False:
                    return 0
                    
        return 1
    
# Time comp:O(N^2)
# Space comp:O(1)
```

```python
"""
1) Create an empty stack.
2) Initialize root as INT_MIN.
3) Do following for every element of an array
     a) If current element is greater than top of the stack then remove all element from stack 
        until top of the stack < current. Ans make last removed element root
     b) If current is smaller than current root, return false.
     c) else push current to stack
"""

class Solution:
    def canRepresentBST(self, arr, N):
        stack = []
        root = float('-inf')
        for i in arr:
            while len(stack) and stack[-1] < i:
                root = stack.pop()
            
            if i < root:
                return 0
            
            stack.append(i)
        return 1
    
# Time comp:O(N)
# Space comp:O(N)
```

### 462. Connect Nodes at Same Level

```python
# Do level order traversal and make a queue
# Traverse a queue and assign next pointer to the next element in queue

class Solution:
    def connect(self, root):
        if (root == None) or (root.left == None and root.right == None):
            return root
        
        q = deque()
        q.append(root)
        q.append(None)
        
        while len(q):
            temp = q.popleft()
            if temp == None:
                break
                
            temp.nextRight = q[0]
            
            if temp.left:
                q.append(temp.left)
            if temp.right:
                q.append(temp.right)
            
            if q[0] == None:
                q.popleft()
                q.append(None)
                
        return root

# Time comp: O(N)
# Auxiliary Space: O(N)   (Because we have use queue here)
```

```python
# Recursive solution
# Traverse each node and assign next nodes of it's children from there only

class Solution:
    def connect(self, root):
        if (root == None) or (root.left == None and root.right == None):
            return root
        
        if root.left:
            if root.right:
                # If right sibling in there then assign it
                root.left.nextRight = root.right
            elif root.nextRight and root.nextRight.left:
                # Else assign left child of right sibling of parent
                root.left.nextRight = root.nextRight.left
            elif root.nextRight and root.nextRight.right:
                # Else assign right child of right sibling of parent
                root.left.nextRight = root.nextRight.right

        if root.right and root.nextRight:
            if root.nextRight and  root.nextRight.left:
                # If left child of right sibling of parent is there then assign it
                root.right.nextRight = root.nextRight.left
            elif root.nextRight and root.nextRight.right:
                # Else assign right child of right sibling of parent
                root.right.nextRight = root.nextRight.right
        
        
        if root.left and (root.left.left != None or root.left.right != None):
            # If left child is not leaf node and make recursive call with it
            self.connect(root.left)
        if root.right and (root.right.left != None or root.right.right != None):
            # If right child is not leaf node and make recursive call with it
            self.connect(root.right)
        
        return root
    
# Time comp: O(N)
# Auxiliary space: O(1)    (recursive stack:O(N))
```

### 464. Sorted array to BST

```python
class Node:
    def __init__(self,k):
        self.data = k
        self.left = None
        self.right = None

class Solution:
    def preorder(self,root,path):
        if root == None:
            return
        
        path.append(root.data)
        self.preorder(root.left,path)
        self.preorder(root.right,path)
    
    # Find middle element, make it root and left of middle as left child and right middle as right child
    def makeBST(self,nums):
        if not nums:
            return None
            
        mid = (len(nums)-1)//2
        root = Node(nums[mid])    # Make mid element as root
        
        root.left = self.makeBST(nums[:mid])
        root.right = self.makeBST(nums[mid+1:])
        
        return root
    
    def sortedArrayToBST(self, nums):
        root = self.makeBST(nums)
        path = []
        self.preorder(root,path)
        return path
    
# Time comp: O(N)
# auxiliary Space: O(N)
```

### 504. Sum of k smallest elements in BST

```python
# Do the inorder traversal and sum first k digits from ans

def solve(root,ans,k):
    if root == None:
        return []
    
    solve(root.left,ans,k)
    ans.append(root.key)
    solve(root.right,ans,k)


def summ(root, k):
    # code here
    ans = []
    solve(root,ans,k)
    
    sum = 0
    for i in range(k):
        sum += ans[i]
    
    return sum

# Time comp:O(N)
# Space comp:O(N)
```
