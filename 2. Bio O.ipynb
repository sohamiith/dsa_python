### Find the time complexity of the following function

```python
# Fine the sum and product of all the elements of an array

def foo(array):
    sum = 0
    product = 1
    for i in array:
        sum += i
    for i in array:
        product *= i
    print("Sum = "+str(sum)+", Product = "+str(product))
    
    
"""
Ans: O(N)
"""
```

```python
# Print all the pairs of an elements of array 

def printPairs(array):
    for i in array:
        for j in array:
            print(str(i)+","+str(j))

            
"""
Ans: O(N^2)
"""
```

```python
# Print unordered pairs from an array

def printUnorderedPairs(array):
    for i in range(0,len(array)):
        for j in range(i+1,len(array)):
            print(array[i] + "," + array[j])
            
"""
Ans: O(N^2)
"""
```

```python
# Print unordered pairs of two array

def printUnorderedPairs(arrayA, arrayB):
    for i in range(len(arrayA)):
        for j in range(len(arrayB)):
            if arrayA[i] < arrayB[j]:
                print(str(arrayA[i]) + "," + str(arrayB[j]))
                
                
"""
Ans: O(N * M)
Where N is size of array A and M is size of array B
"""
```

```python
# Reverse

def reverse(array):
    for i in range(0,int(len(array)/2)):
        other = len(array)-i-1
        temp = array[i]
        array[i] = array[other]
        array[other] = temp
    print(array)
    
"""
Ans: O(N)
"""
```

```python
# Find factorial using recursion

def factorial(n):
    if n < 0:
        return -1
    elif n == 0:
        return 1
    else:
        return n * factorial(n-1)
    
"""
Ans: O(N)
"""
```

```python
# Fibonacci of the given numbers

def fib(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    return fib(n-1) + fib(n-2)

"""
Ans: O(2^N) 
"""

# Fibonacci of all the numbers

def allFib(n):
    for i in range(n):
        print(str(i)+":, " + str(fib(i)))

"""
Ans: O(2^N)
"""
        
```
